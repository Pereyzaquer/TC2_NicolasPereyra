```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode

numerator = [3,0,7,0]
denominator = [1, 0, 7, 0, 10]
system = TransferFunction(numerator, denominator)

freq = np.logspace(-1, 2, 1000)
omega, mag, phase = bode(system, freq)

plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(omega, mag)
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Magnitud [dB]')
plt.grid()

plt.subplot(2, 1, 2)
plt.semilogx(omega, phase)
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Fase [grados]')
plt.grid()
plt.xlim(0.01,100)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import TransferFunction, bode
k = 1/2
numerator = [k,k]
denominator = [1, 6,8]
system = TransferFunction(numerator, denominator)

freq = np.logspace(-1, 2, 1000)
omega, mag, phase = bode(system, freq)

plt.figure()
plt.subplot(2, 1, 1)
plt.semilogx(omega, mag)
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Magnitud [dB]')
plt.grid()

plt.subplot(2, 1, 2)
plt.semilogx(omega, phase)
plt.xlabel('Frecuencia [rad/s]')
plt.ylabel('Fase [grados]')
plt.grid()
plt.xlim(0.01,100)

plt.tight_layout()
plt.show()
```
